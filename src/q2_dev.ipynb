## Solución 2

Se lee el archivo con la libreria JSON y luego se toma solo la columna content, que tiene el texto, para ver si hay emojis en el texto.

````python
import json
import polars as pl

file_path = "farmers-protest-tweets-2021-2-4.json"

dict_data = {}
content_list = []
with open(file_path) as json_file:
    for line in json_file.readlines():
        json_obj = json.loads(line)
        content_list.append(json_obj['content'])

dict_data['content'] = content_list
````

Se utiliza la misma función previa para encontrar emojis.

````python
emojis = map(get_emojis_in_content, dict_data['content'])
````

Se filtran los elementos de la lista que tengan valor None.

````python
emojis_filtered = filter(lambda x: x is not None, emojis)
````

Se usa el método `chain.from_iterable` de la librería `itertools` para aplanar la lista, dado que tenemos una lista de listas, y al aplanarla se facilitará contar la cantidad de veces que aparece cada emoji. 

````python
from itertools import chain


emojis_full = chain.from_iterable(emojis_filtered)
````

Se utiliza un objeto Counter para contar la cantidad de veces que aparece cada emoji en la lista.

````python
from collections import Counter

emojis_counter = Counter(emojis_full)
emojis_counter
````

Se seleccionan los 10 emojis que aparecen más veces.

````python
top_10_emojis = sorted(emojis_counter.items(), key=lambda x: x[1], reverse=True)[:10]
top_10_emojis
````

Resumen

````python
import json
import polars as pl
from collections import Counter
from itertools import chain
import emoji

def get_emojis_in_content(x):
    result =  [x['emoji'] for x in emoji.emoji_list(x)]
    return result if len(result) > 0 else None

file_path = "farmers-protest-tweets-2021-2-4.json"

dict_data = {}
ids = []
content_list = []
with open(file_path) as json_file:
    for line in json_file.readlines():
        json_obj = json.loads(line)
        ids.append(json_obj['id'])
        content_list.append(json_obj['content'])

dict_data['id'] = ids
dict_data['content'] = content_list

emojis = map(get_emojis_in_content, dict_data['content'])
emojis_filtered = filter(lambda x: x is not None, emojis)
emojis_full = chain.from_iterable(emojis_filtered)
emojis_counter = Counter(emojis_full)
top_10_emojis = sorted(emojis_counter.items(), key=lambda x: x[1], reverse=True)[:10]
top_10_emojis
````
