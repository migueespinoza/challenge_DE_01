## Q3 Memory

Se lee el archivo con la libreria json y se extrae solo la columna content, que tiene el texto y los usuarios mencionados.

````python
import json

dict_data = {}
content_list = []
with open(file_path) as json_file:
    for line in json_file.readlines():
        json_obj = json.loads(line)
        content_list.append(json_obj['content'])

dict_data['content'] = content_list
````

Se extrae el usuario mencionado a partir del texto.

````python
import re
from itertools import chain
from collections import Counter

def find_mentioned_users(text):
    result = re.findall('@(\w+)', text)
    return result if len(result) > 0 else None

mentioned_users = list(map(find_mentioned_users, dict_data['content']))
````

Se limpia la lista de usuarios mencionados. En primer lugar se eliminan los None, y luego se unen las listas.

````python
mentioned_users_filtered = filter(lambda x: x is not None, mentioned_users)
mentioned_users_full = chain.from_iterable(mentioned_users_filtered)
mentioned_users_full
````

Se utiliza un objeto Counter para contar los elementos de la lista. Luego se seleccionan los 10 primeros por cantidad.

````python
users_counter = Counter(mentioned_users_full)
top_10_user_mentioned = sorted(users_counter.items(), key=lambda x: x[1], reverse=True)[:10]
top_10_user_mentioned
````

Resumen

````python
import json
import re
from itertools import chain
from collections import Counter

def find_mentioned_users(text):
    result = re.findall('@(\w+)', text)
    return result if len(result) > 0 else None

file_path = "farmers-protest-tweets-2021-2-4.json"
dict_data = {}
content_list = []
with open(file_path) as json_file:
    for line in json_file.readlines():
        json_obj = json.loads(line)
        content_list.append(json_obj['content'])

dict_data['content'] = content_list
mentioned_users = list(map(find_mentioned_users, dict_data['content']))
mentioned_users_filtered = filter(lambda x: x is not None, mentioned_users)
mentioned_users_full = chain.from_iterable(mentioned_users_filtered)
users_counter = Counter(mentioned_users_full)
top_10_user_mentioned = sorted(users_counter.items(), key=lambda x: x[1], reverse=True)[:10]
````
