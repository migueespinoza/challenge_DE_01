```python
import json

path_tweets = 'farmers-protest-tweets-2021-2-4.json'
with open(path_tweets) as t:
    tweets = json.load(t)

tweets
```

Esto ocurre porque el archivo no es un json como tal, ya que un json contiene solo un diccionario enterno, en cambio este archivo tiene muchos diccionarios externos todos al mismo nivel. Por lo que para trabajar con Ã©l hay que leer linea a linea, y decodificar el json.

```python
json_lines = []
with open(path_tweets) as t:
    for i, line in enumerate(t):
        json_lines.append(line)
        if i >= 5:
            break
json_lines
```

```python
import pandas as pd

df_tweets = pd.read_json(path_tweets, lines = True, nrows = 5)
df_tweets
```
