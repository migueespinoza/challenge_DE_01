Primero se har√° esta parte: "Las top 10 fechas donde hay m√°s tweets."

Soluci√≥n:
La fecha viene en la columna date. Por lo tanto se debe hacer un groupby por fecha y contar la cantidad de tweets. Luego ordenar por cantidad descendentemente, y sacar las 10 primeras fechas, que ser√≠an las con m√°s tweets.

````python
import pandas as pd

path_tweets = 'farmers-protest-tweets-2021-2-4.json'
df_tweets_full = pd.read_json(path_tweets, lines = True)
print(f'Cantidad de filas: {len(df_tweets_full)}')
print(f'Cantidad de URLs unicas: {len(df_tweets_full["url"].unique())}')
df_tweets_full.dtypes
````

````python
pd.options.display.max_colwidth = None
pd.options.display.max_columns = None
df_tweets_full.head(3)
````

Se pasar√° la columna "date" a tipo date, es decir, se eliminara la parte horaria.

````python
df_tweets_full['date_only'] = df_tweets_full['date'].dt.date
df_tweets_full[['date', 'date_only']]
````

````python
cantidad_url_por_fecha = df_tweets_full.groupby('date_only')['url'].count().reset_index()
cantidad_url_por_fecha
````

````python
top10_fechas = cantidad_url_por_fecha.rename(columns={'url':'count'}).sort_values(by='count', ascending=False).head(10)
top10_fechas
````

````python
top_fechas = top10_fechas['date_only'].to_list()
````

````python
top_fechas
````

Ahora que se tienen las fechas, hay que buscarlas en el archivo original de tweets, ya que se quiere saber la cantidad de usuarios con mayor cantidad de tweets en esas fechas.

````python
df_tweets_top_fechas = df_tweets_full[df_tweets_full['date_only'].isin(top_fechas)]
df_tweets_top_fechas.head(2)
````

Aqui debo sacar el username de la columna user.

````python
df_tweets_top_fechas['user']
````

````python
df_tweets_top_fechas['user'].iloc[0]
````

````python
df_tweets_top_fechas['username'] = df_tweets_top_fechas['user'].apply(lambda x: x['username'])
df_tweets_top_fechas['username']
````

Ahora que se tienen los users, se puede contar la cantidad de posts que hizo cada user en esas fechas.

````python
cantidad_tweets_por_fecha_y_user = df_tweets_top_fechas.groupby(by=['date_only', 'username'], sort=False)['url'].count().reset_index().\
    rename(columns={'url': 'count'})
cantidad_tweets_por_fecha_y_user
````

````python
mayor_cantidad_tweets_de_un_usuario_por_fecha = cantidad_tweets_por_fecha_y_user.\
    groupby(by=['date_only'], sort=False)['count'].agg(['max', 'sum']).reset_index()
mayor_cantidad_tweets_de_un_usuario_por_fecha
````

````python
join_dfs = cantidad_tweets_por_fecha_y_user.merge(mayor_cantidad_tweets_de_un_usuario_por_fecha, on='date_only', how='inner')
usuario_max_tweets_por_fecha = join_dfs[join_dfs['count'] == join_dfs['max']].reset_index(drop=True).sort_values(by='sum', ascending=False)
usuario_max_tweets_por_fecha
````

````python
lista_usuario_max_tweets_por_fecha = [(x, y) for x, y in usuario_max_tweets_por_fecha[['date_only', 'username']].values]
lista_usuario_max_tweets_por_fecha
````
